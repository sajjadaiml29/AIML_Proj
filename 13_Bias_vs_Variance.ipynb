## What is Error?

In Machine Learning, the goal is to build a model that makes accurate predictions.

The difference between the actual value and the predicted value is called the **prediction error**.

Bias and Variance are two major sources of prediction error.

## What is Bias?

Bias is the error caused by making overly simple assumptions when building a Machine Learning model.

A model with high bias cannot learn the underlying patterns in the data.

High Bias usually leads to **Underfitting**.

## What is Variance?

Variance measures how much a Machine Learning model changes when it is trained on different datasets.

A model with high variance learns the training data extremely well, including its noise.

High Variance usually leads to **Overfitting**.

## Underfitting

Underfitting occurs when the model is too simple to capture the patterns in the data.

Characteristics:

- High Bias
- Low Variance
- Poor training accuracy
- Poor testing accuracy

## Overfitting

Overfitting occurs when the model learns the training data too well, including noise and random variations.

Characteristics:

- Low Bias
- High Variance
- Excellent training accuracy
- Poor testing accuracy

## Good Fit

A good Machine Learning model balances Bias and Variance.

Characteristics:

- Low Bias
- Low Variance
- Good training accuracy
- Good testing accuracy

## Bias-Variance Tradeoff

The Bias-Variance Tradeoff refers to finding the right balance between Bias and Variance.

Reducing Bias usually increases Variance.

Reducing Variance usually increases Bias.

The objective is to find a model that generalizes well to unseen data.

# Explain
Training error always decreases.

Testing error decreases then increases.

The best model is in the middle.




```python
#Underfitting vs Overfitting
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 200)

training_error = np.exp(-0.4 * x) + 0.05

testing_error = 0.05 * (x - 5) ** 2 + 0.35

plt.figure(figsize=(8,5))

plt.plot(
    x,
    training_error,
    label="Training Error",
    linewidth=3
)

plt.plot(
    x,
    testing_error,
    label="Testing Error",
    linewidth=3
)

plt.xlabel("Model Complexity")

plt.ylabel("Error")

plt.title("Bias-Variance Tradeoff")

plt.legend()

plt.grid(True)

plt.show()
```

## Real-World Example

Imagine a teacher preparing students for an exam.

Underfitting:

The teacher only explains the basic concepts.

Students perform poorly because they have not learned enough.

Overfitting:

The teacher makes students memorize last year's exam questions.

Students perform well only if the same questions appear again.

Good Fit:

The teacher explains the concepts and provides enough practice.

Students can solve both familiar and new questions successfully.

## Comparison Table

| High Bias              | High Variance               |
| ---------------------- | --------------------------- |
| Underfitting           | Overfitting                 |
| Model too simple       | Model too complex           |
| Misses patterns        | Learns noise                |
| Poor training accuracy | Excellent training accuracy |
| Poor testing accuracy  | Poor testing accuracy       |


## Which Algorithms Usually Have High Bias?

| Algorithm           | Bias |
| ------------------- | ---- |
| Linear Regression   | High |
| Logistic Regression | High |
| Naive Bayes         | High |


## Which Algorithms Usually Have High Variance?

| Algorithm     | Variance |
| ------------- | -------- |
| Decision Tree | High     |
| Random Forest | Low      |
| Bagging       | Low      |
| Ridge         | Lower    |
| Lasso         | Lower    |


## How to Reduce Bias

Increase model complexity.

Examples:

- Add more features.
- Use a more complex algorithm.
- Reduce regularization.

## How to Reduce Variance

Reduce model complexity.

Examples:

- Collect more training data.
- Use Bagging.
- Use Random Forest.
- Apply Regularization.
- Perform Cross Validation.

## Final Summary

| Problem      | Cause         | Solution            |
| ------------ | ------------- | ------------------- |
| Underfitting | High Bias     | Increase complexity |
| Overfitting  | High Variance | Reduce complexity   |


# Conclusion

Bias and Variance are two important concepts in Machine Learning that directly affect a model's performance.

High Bias leads to Underfitting, while High Variance leads to Overfitting.

The goal is to build a model that balances Bias and Variance, allowing it to generalize well to new, unseen data.
