# Decision Tree Regression

## Objective

The objective of this notebook is to build a Decision Tree Regression model that predicts an employee's monthly salary.

In this notebook, you will learn how to:

- Load the dataset
- Handle missing values
- Define features and target variable
- Split the dataset into training and testing sets
- Train a Decision Tree Regression model
- Make predictions
- Evaluate the model using regression metrics
- Visualize the Decision Tree

```python
# Import Required Libraries

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.tree import (
    DecisionTreeRegressor,
    plot_tree
)

from sklearn.metrics import (
    r2_score,
    mean_absolute_error,
    mean_squared_error
)

import matplotlib.pyplot as plt
```

**Step 1: Mount Google Drive**

Google Drive is mounted to access the dataset and save the trained model.

```python
from google.colab import drive

drive.mount('/content/drive')
```

**Step 2: Load the Dataset**

Load the Employee Salary Classification dataset from Google Drive.

```python
df = pd.read_excel(
    "/content/drive/MyDrive/Datasets/Salary_Logistic_Regression_Multiclass_Classification_Dataset.xlsx"
)

df
```

**Step 3: Handle Missing Values**

Check whether the dataset contains missing values.

If any missing values are found, replace them with the median value of the respective column.

```python
print("\nBefore Filling Missing Values")
print(df.isnull().sum())

df["Education_Level"] = df["Education_Level"].fillna(
    df["Education_Level"].median()
)

df["Age"] = df["Age"].fillna(
    df["Age"].median()
)

print("\nAfter Filling Missing Values")

print(df.isnull().sum())
```

**Step 4: Define Features and Target Variable**

The independent variables are:

- Experience Years
- Education Level
- Age

The dependent variable is:

- Monthly Salary (PKR)

```python
X = df[
    [
        "Experience_Years",
        "Education_Level",
        "Age"
    ]
]

y = df["Monthly_Salary_PKR"]
```

**Step 5: Split the Dataset**

Split the dataset into training and testing sets.

- 80% Training Data
- 20% Testing Data

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

**Step 6: Train the Decision Tree Regression Model**

Train the Decision Tree Regression model using the training dataset.

A Decision Tree predicts numerical values by creating a series of decision rules based on the input features.

A maximum tree depth of **3** is used to keep the model simple and reduce overfitting.

```python
model = DecisionTreeRegressor(
    max_depth=3,
    random_state=42
)

model.fit(
    X_train,
    y_train
)

```

**Step 7: Make Predictions**

Predict the monthly salary using the testing dataset.

```python
y_pred = model.predict(
    X_test
)

print(y_pred)
```

**Step 8: Evaluate Model Performance**

The model performance is evaluated using the following metrics:

- **R² Score:** Measures how well the model explains the variation in the target variable.
- **MAE (Mean Absolute Error):** Measures the average absolute difference between the actual and predicted values.
- **MSE (Mean Squared Error):** Measures the average squared difference between the actual and predicted values.
- **RMSE (Root Mean Squared Error):** Measures the average prediction error in the same unit as the target variable.

```python
r2 = r2_score(
    y_test,
    y_pred
)

mae = mean_absolute_error(
    y_test,
    y_pred
)

mse = mean_squared_error(
    y_test,
    y_pred
)

rmse = np.sqrt(mse)

print("R² Score :", round(r2,4)*100)

print("MAE :", round(mae,2))

print("MSE :", round(mse,2))

print("RMSE :", round(rmse,2))
```

**Step 9: Predict the Monthly Salary for a New Employee**

Predict the monthly salary for an employee with:

- Experience = 10 Years
- Education Level = 18 Years
- Age = 35 Years

```python
prediction = model.predict(
    [[10,18,35]]
)

print("Predicted Monthly Salary:")

print(
    round(prediction[0],2),
    "PKR"
)
```

**Step 10: Visualize the Decision Tree**

Visualize the trained Decision Tree model to understand how it predicts numerical values.

Each node represents a decision based on a feature value, and each leaf node represents the predicted monthly salary.

```python

plt.figure(figsize=(22,12), dpi=150)

plot_tree(
    model,
    feature_names=X.columns,
    filled=True,
    rounded=True,
    fontsize=10,
    precision=2,
    impurity=True,
    proportion=True,
    node_ids=True
)

plt.title(
    "Decision Tree Regression",
    fontsize=18,
    fontweight="bold"
)

plt.tight_layout()

plt.show()

print("Decision Tree Generated Successfully")
```

**Conclusion**

In this notebook, a Decision Tree Regression model was successfully developed to predict employees' monthly salaries. The dataset was preprocessed by handling missing values and splitting it into training and testing sets. The trained model was then used to predict salaries for new employees. Finally, the model's performance was evaluated using R² Score, MAE, MSE, and RMSE, and the Decision Tree was visualized to understand how it makes predictions.
