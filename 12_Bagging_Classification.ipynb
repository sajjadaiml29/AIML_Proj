# Bagging Classification

## Objective

The objective of this notebook is to build a Bagging Classification model that predicts an employee's salary category.

In this notebook, you will learn how to:

- Load the dataset
- Handle missing values
- Define features and target variable
- Split the dataset into training and testing sets
- Scale the features
- Train a Decision Tree model
- Train a Bagging Classification model
- Compare the performance of both models

```python
# Import Required Libraries

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler

from sklearn.tree import DecisionTreeClassifier

from sklearn.ensemble import BaggingClassifier

from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report
)
```

**Step 1: Mount Google Drive**

Google Drive is mounted to access the dataset and save the trained model.

```python
from google.colab import drive

drive.mount('/content/drive')
```

**Step 2: Load the Dataset**

Load the Employee Salary Classification dataset from Google Drive.

```python
df = pd.read_excel(
    "/content/drive/MyDrive/Datasets/Salary_Logistic_Regression_Multiclass_Classification_Dataset.xlsx"
)

df
```

## Step 3: Handle Missing Values

Check whether the dataset contains missing values.

If any missing values are found, replace them with the median value of the respective column.

```python
print("\nBefore Filling Missing Values")
print(df.isnull().sum())

df["Education_Level"] = df["Education_Level"].fillna(
    df["Education_Level"].median()
)

df["Age"] = df["Age"].fillna(
    df["Age"].median()
)

print("\nAfter Filling Missing Values")

print(df.isnull().sum())
```

## Step 4: Define Features and Target Variable

The independent variables are:

- Experience Years
- Education Level
- Age

The dependent variable is:

- Salary Category

Salary Categories:

- 0 = Low Salary
- 1 = Medium Salary
- 2 = High Salary
- 3 = Executive Salary

```python
X = df[
    [
        "Experience_Years",
        "Education_Level",
        "Age"
    ]
]

y = df["Salary_Category"]
```

## Step 5: Split the Dataset

Split the dataset into training and testing sets.

- 80% Training Data
- 20% Testing Data

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

## Step 6: Scale the Features

Standardize the feature values so that each feature has a similar scale.

Feature scaling helps improve the performance of the Bagging Classification model.

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(
    X_train
)

X_test = scaler.transform(
    X_test
)
```

## Step 7: Train the Decision Tree and Bagging Classification Models

Train a Decision Tree model and then use it as the base estimator for the Bagging Classification model.

The **n_estimators** parameter specifies the number of Decision Trees used in the Bagging model.

```python
decision_tree = DecisionTreeClassifier(
    random_state=42
)

bagging_model = BaggingClassifier(
    estimator=decision_tree,
    n_estimators=100,
    random_state=42
)

bagging_model.fit(
    X_train,
    y_train
)

```

## Step 8: Make Predictions

Use both the Decision Tree model and the Bagging Classification model to predict the salary categories for the testing dataset.

```python
# Train the Decision Tree Model

decision_tree.fit(
    X_train,
    y_train
)

# Make Predictions

decision_tree_predictions = decision_tree.predict(
    X_test
)

bagging_predictions = bagging_model.predict(
    X_test
)
```

## Step 9: Compare Model Accuracy

Calculate and compare the accuracy of the Decision Tree model and the Bagging Classification model.

This comparison helps determine whether Bagging improves the model's performance.

```python
decision_tree_accuracy = accuracy_score(
    y_test,
    decision_tree_predictions
)

bagging_accuracy = accuracy_score(
    y_test,
    bagging_predictions
)

print("Decision Tree Accuracy:")

print(
    round(decision_tree_accuracy * 100,2),
    "%"
)

print()

print("Bagging Accuracy:")

print(
    round(bagging_accuracy * 100,2),
    "%"
)
```

## Step 10: Display the Confusion Matrix

Display the confusion matrix for the Bagging Classification model.

The confusion matrix compares the actual salary categories with the predicted salary categories.

```python
cm = confusion_matrix(
    y_test,
    bagging_predictions
)

print("Confusion Matrix:")

print(cm)
```

## Step 11: Display the Classification Report

Display the classification report for the Bagging Classification model.

The report includes:

- Precision
- Recall
- F1-Score
- Support

```python
print("Classification Report:")

print()

print(
    classification_report(
        y_test,
        bagging_predictions
    )
)
```

## Step 12: Predict the Salary Category for a New Employee

Predict the salary category for an employee with:

- Experience = 12 Years
- Education Level = 20 Years
- Age = 40 Years

```python
new_employee = scaler.transform(
    [[12,20,40]]
)

prediction = bagging_model.predict(
    new_employee
)

salary_categories = {
    0: "Low Salary",
    1: "Medium Salary",
    2: "High Salary",
    3: "Executive Salary"
}

print("Predicted Salary Category:")

print(
    salary_categories[prediction[0]]
)
```

A Decision Tree was used as the base estimator, and multiple Decision Trees were combined using the Bagging technique. The performance of the Bagging model was compared with a single Decision Tree model, demonstrating how Bagging can improve prediction accuracy and reduce overfitting.
