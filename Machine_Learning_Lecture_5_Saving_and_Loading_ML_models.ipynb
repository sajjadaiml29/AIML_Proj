**Saving and Loading a Machine Learning Model**

**Objective**

The objective of this project is to learn how to save a trained Machine Learning model and reload it later for making predictions without retraining.

In this notebook, we will:

- Train a Multiple Linear Regression model
- Save the trained model using Pickle
- Reload the saved model
- Make predictions using the loaded model
- Save the model using Joblib
- Reload the Joblib model
- Verify that both saved models produce the same predictions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

import pickle
import joblib
```

**Step 1: Mount Google Drive**

Google Drive is mounted to access the dataset and save the trained model.

```python
from google.colab import drive

drive.mount('/content/drive')
```

**Step 2: Load the Dataset**

Load the employee salary dataset from Google Drive.

```python
df = pd.read_excel(
    "/content/drive/MyDrive/Datasets/Salary_Multiple_Linear_Regression_Dataset.xlsx"
)

df
```

**Step 3: Handle Missing Values**

Check for missing values in the dataset.

Replace missing values with the median of the corresponding column.

```python
print("\nBefore Filling Missing Values")
print(df.isnull().sum())

df["Education_Level"] = df["Education_Level"].fillna(
    df["Education_Level"].median()
)

df["Age"] = df["Age"].fillna(
    df["Age"].median()
)

print("\nAfter Filling Missing Values")

print(df.isnull().sum())
```

**Step 4: Define Features and Target Variable**

Independent Variables (X):

- Experience_Years
- Education_Level
- Age

Dependent Variable (Y):

- Monthly_Salary_PKR

```python
X = df[
    [
        "Experience_Years",
        "Education_Level",
        "Age"
    ]
]

y = df["Monthly_Salary_PKR"]

print(X.head())
print(y.head())
```

**Step 5: Train the Multiple Linear Regression Model**

Train the model using the employee salary dataset.

After training, the model can be used for salary prediction.

```python
model = LinearRegression()

model.fit(X, y)

print("Model Trained Successfully")
```

**Step 6: Save the Model Using Pickle**

Pickle is a Python module used to save trained machine learning models.

The model is saved in binary format using `wb`, which means write binary.

```python
with open(
    "/content/drive/MyDrive/Models/Salary_Multiple_Linear_Regression_Model.pkl","wb"
) as file:

    pickle.dump(model, file)

print("Model Saved Successfully Using Pickle")
```

**Step 7: Load the Saved Pickle Model**

Now we load the saved model from Google Drive.

The model is loaded using `rb`, which means read binary.

```python
with open(
    "/content/drive/MyDrive/Models/Salary_Multiple_Linear_Regression_Model.pkl","rb"
) as file:

    pickle_model = pickle.load(file)

print("Pickle Model Loaded Successfully")
```

**Step 8: Make Prediction Using Pickle Model**

Now we test the loaded Pickle model by predicting salary for a new employee.

Input values:

- Experience Years = 10
- Education Level = 18
- Age = 35

```python
pickle_prediction = pickle_model.predict(
    [[10, 18, 35]]
)

print("Predicted Salary using Pickle Model:")

print(round(pickle_prediction[0], 2), "PKR")
```

**Step 9: Save the Model Using Joblib**

Joblib is another method used to save machine learning models.

It is commonly used for models that contain large NumPy arrays.

```python
joblib.dump(
    model,
    "/content/drive/MyDrive/Models/Salary_Multiple_Linear_Regression_Model.joblib"
)

print("Model Saved Successfully Using Joblib")
```

**Step 10: Load the Saved Joblib Model**

Load the trained model that was saved using Joblib.

This model can now be used to make predictions without retraining.

```python
loaded_joblib_model = joblib.load(
    "/content/drive/MyDrive/Models/Salary_Multiple_Linear_Regression_Model.joblib"
)

print("Joblib Model Loaded Successfully")
```

**Step 11: Make Prediction Using Joblib Model**

Use the loaded Joblib model to predict the monthly salary of an employee.

Input values:

- Experience Years = 10
- Education Level = 18
- Age = 35

```python
joblib_prediction = loaded_joblib_model.predict(
    [[10, 18, 35]]
)

print("Predicted Salary using Joblib Model:")

print(round(joblib_prediction[0], 2), "PKR")
```

**Step 12: Compare Predictions**

Compare the prediction results from the Pickle model and the Joblib model.

Since both models were saved from the same trained model, they should produce identical predictions.

```python
print("Prediction using Pickle :", round(pickle_prediction[0], 2), "PKR")

print("Prediction using Joblib:", round(joblib_prediction[0], 2), "PKR")
```

**Step 13: Verify Both Predictions**

Check whether both models return the same prediction result.

```python
if round(pickle_prediction[0], 2) == round(joblib_prediction[0], 2):
    print("✅ Both models produced the same prediction.")
else:
    print("❌ The predictions are different.")
```

**Conclusion**

In this notebook, a Multiple Linear Regression model was successfully trained using the employee salary dataset.

The trained model was saved using two different methods:

- Pickle
- Joblib

Both saved models were loaded successfully and produced the same salary prediction. This demonstrates that trained machine learning models can be saved, shared, and reused without retraining, which is an important step before deploying models in real-world applications.
