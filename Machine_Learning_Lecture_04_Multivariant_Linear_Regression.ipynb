**Multiple Linear Regression for Salary Prediction**

**Objective**

The objective of this project is to predict monthly salary using multiple factors such as experience, education level, and age.

This project uses Multiple Linear Regression because there are multiple input features affecting the salary.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

**Step 1: Load Dataset**

The dataset contains three independent variables:

- Experience_Years
- Education_Level
- Age

The target variable is:

- Monthly_Salary_PKR

```python
df = pd.read_excel(
    "/content/drive/MyDrive/Datasets/Salary_Multiple_Linear_Regression_Dataset.xlsx"
)

df
```

**Step 2: Check Dataset Information**

Before training the model, we examine the dataset structure and check for missing values.

```python
print("Dataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nMissing Values:")
print(df.isnull().sum())

print("\nStatistical Summary:")
df.describe()
```

**Step 3: Handle Missing Values**

The dataset contains missing values.

These missing values are replaced using the median value of the corresponding column.

```python
df["Education_Level"] = df["Education_Level"].fillna(
    df["Education_Level"].median()
)

df["Age"] = df["Age"].fillna(
    df["Age"].median()
)

print(df.isnull().sum())
```

**Step 4: Define Features and Target Variable**

Independent Variables (X):

- Experience_Years
- Education_Level
- Age

Dependent Variable (Y):

- Monthly_Salary_PKR

```python
X = df[
    [
        "Experience_Years",
        "Education_Level",
        "Age"
    ]
]

y = df["Monthly_Salary_PKR"]

print(X.head())
print(y.head())
```

**Step 5: Train Multiple Linear Regression Model**

The model is trained using employee data.

```python
model = LinearRegression()

model.fit(X, y)
```

**Step 6: Display Coefficients and Intercept**

The model calculates coefficients for each independent variable and an intercept value.

```python
print("Coefficients:")

for column, coefficient in zip(X.columns, model.coef_):
    print(f"{column}: {coefficient}")

print("\nIntercept:")
print(model.intercept_)
```

**Step 7: Regression Equation**

The Multiple Linear Regression equation is:

Salary = b0 + b1(Experience) + b2(Education) + b3(Age)

Where:

- b0 = Intercept
- b1, b2, b3 = Coefficients

```python
print("Regression Equation:\n")

print(
    f"Salary = {model.intercept_:.2f}"
    f" + ({model.coef_[0]:.2f} × Experience)"
    f" + ({model.coef_[1]:.2f} × Education)"
    f" + ({model.coef_[2]:.2f} × Age)"
)
```

**Step 8: Predict Salary**

Predict salary for:

- Experience = 10 Years
- Education Level = 18 Years
- Age = 35 Years

```python
predicted_salary = model.predict(
    [[10, 18, 35]]
)

print(
    "Predicted Salary:",
    round(predicted_salary[0], 2),
    "PKR"
)
```

**Step 9: Multiple Employee Predictions**

The model predicts salaries for multiple employees.

```python
future_employees = pd.DataFrame({
    "Experience_Years":[2,4,6,8,10],
    "Education_Level":[14,16,16,18,18],
    "Age":[24,28,32,35,38]
})

future_employees["Predicted_Salary"] = model.predict(
    future_employees
)

future_employees
```

**Step 10: Evaluate Model Performance**

The model performance is measured using:

- R² Score
- MAE
- MSE
- RMSE

```python
predictions = model.predict(X)

r2 = r2_score(y, predictions)

mae = mean_absolute_error(y, predictions)

mse = mean_squared_error(y, predictions)

rmse = np.sqrt(mse)

print(f"R² Score : {r2*100:.2f}%")
print(f"MAE      : {mae:.2f}")
print(f"MSE      : {mse:.2f}")
print(f"RMSE     : {rmse:.2f}")
```

**Conclusion**

A Multiple Linear Regression model was successfully developed to predict employee salaries using Experience Years, Education Level, and Age.

Missing values were handled using median imputation, and the model achieved a strong prediction performance. The trained model can be used to estimate salaries for new employees based on their characteristics.
