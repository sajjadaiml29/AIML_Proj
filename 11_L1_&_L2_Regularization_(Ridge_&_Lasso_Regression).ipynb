# L1 & L2 Regularization (Ridge & Lasso Regression)

## Objective

The objective of this notebook is to improve a Linear Regression model using L1 and L2 Regularization.

In this notebook, you will learn how to:

- Load the dataset
- Handle missing values
- Define features and target variable
- Split the dataset into training and testing sets
- Scale the features
- Train a Ridge Regression model
- Train a Lasso Regression model
- Compare both models

```python
# Import Required Libraries

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import (
    Ridge,
    Lasso
)

from sklearn.metrics import (
    r2_score,
    mean_absolute_error,
    mean_squared_error
)
```

**Step 1: Mount Google Drive**

Google Drive is mounted to access the dataset and save the trained model.

```python
from google.colab import drive

drive.mount('/content/drive')
```

**Step 2: Load the Dataset**

Load the Employee Salary Classification dataset from Google Drive.

```python
df = pd.read_excel(
    "/content/drive/MyDrive/Datasets/Salary_Logistic_Regression_Multiclass_Classification_Dataset.xlsx"
)

df
```

## Step 3: Handle Missing Values

Check whether the dataset contains missing values.

If any missing values are found, replace them with the median value of the respective column.

```python
print("\nBefore Filling Missing Values")
print(df.isnull().sum())

df["Education_Level"] = df["Education_Level"].fillna(
    df["Education_Level"].median()
)

df["Age"] = df["Age"].fillna(
    df["Age"].median()
)

print("\nAfter Filling Missing Values")

print(df.isnull().sum())
```

## Step 4: Define Features and Target Variable

The independent variables are:

- Experience Years
- Education Level
- Age

The dependent variable is:

- Monthly Salary (PKR)

```python
X = df[
    [
        "Experience_Years",
        "Education_Level",
        "Age"
    ]
]

y = df["Monthly_Salary_PKR"]
```

## Step 5: Split the Dataset

Split the dataset into training and testing sets.

- 80% Training Data
- 20% Testing Data

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

## Step 6: Scale the Features

Standardize the feature values so that each feature has a similar scale.

Feature scaling improves the performance of Ridge and Lasso Regression because both algorithms are sensitive to the magnitude of the feature values.

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(
    X_train
)

X_test = scaler.transform(
    X_test
)
```

## Step 7: Train the Ridge Regression Model

Train the Ridge Regression model using the training dataset.

The **alpha** parameter controls the amount of L2 Regularization applied to the model.

```python
ridge_model = Ridge(
    alpha=1.0
)

ridge_model.fit(
    X_train,
    y_train
)

print("Ridge Regression Model Trained Successfully")
```

## Step 8: Train the Lasso Regression Model

Train the Lasso Regression model using the training dataset.

The **alpha** parameter controls the amount of L1 Regularization applied to the model.

```python
lasso_model = Lasso(
    alpha=1.0
)

lasso_model.fit(
    X_train,
    y_train
)

print("Lasso Regression Model Trained Successfully")
```

## Step 9: Make Predictions

Use both the Ridge Regression model and the Lasso Regression model to predict the monthly salaries for the testing dataset.

```python
ridge_predictions = ridge_model.predict(
    X_test
)

lasso_predictions = lasso_model.predict(
    X_test
)
```

## Step 10: Evaluate the Ridge Regression Model

Evaluate the Ridge Regression model using:

- R² Score
- MAE
- MSE
- RMSE

```python
ridge_r2 = r2_score(
    y_test,
    ridge_predictions
)

ridge_mae = mean_absolute_error(
    y_test,
    ridge_predictions
)

ridge_mse = mean_squared_error(
    y_test,
    ridge_predictions
)

ridge_rmse = np.sqrt(
    ridge_mse
)

print("Ridge Regression Performance")

print("----------------------------")

print("R² Score :", round(ridge_r2,4))

print("MAE :", round(ridge_mae,2))

print("MSE :", round(ridge_mse,2))

print("RMSE :", round(ridge_rmse,2))
```

## Step 11: Evaluate the Lasso Regression Model

Evaluate the Lasso Regression model using:

- R² Score
- MAE
- MSE
- RMSE

```python
lasso_r2 = r2_score(
    y_test,
    lasso_predictions
)

lasso_mae = mean_absolute_error(
    y_test,
    lasso_predictions
)

lasso_mse = mean_squared_error(
    y_test,
    lasso_predictions
)

lasso_rmse = np.sqrt(
    lasso_mse
)

print("Lasso Regression Performance")

print("----------------------------")

print("R² Score :", round(lasso_r2,4))

print("MAE :", round(lasso_mae,2))

print("MSE :", round(lasso_mse,2))

print("RMSE :", round(lasso_rmse,2))
```

## Step 12: Compare the Ridge and Lasso Regression Models

Compare the performance of both models and predict the monthly salary for a new employee with:

- Experience = 10 Years
- Education Level = 18 Years
- Age = 35 Years

```python
new_employee = scaler.transform(
    [[10,18,35]]
)

ridge_salary = ridge_model.predict(
    new_employee
)

lasso_salary = lasso_model.predict(
    new_employee
)

print("Predicted Monthly Salary")

print("------------------------")

print(
    "Ridge Regression :",
    round(ridge_salary[0],2),
    "PKR"
)

print(
    "Lasso Regression :",
    round(lasso_salary[0],2),
    "PKR"
)
```
