**Logistic Regression (Binary Classification)**

**Objective**

The objective of this notebook is to build a Logistic Regression model that predicts whether an employee belongs to the High Salary category.

In this notebook, you will learn how to:

- Load a classification dataset
- Handle missing values
- Define features and target variable
- Split the dataset into training and testing sets
- Scale numerical features
- Train a Logistic Regression model
- Make predictions
- Evaluate the model using classification metrics

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression

from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report
)
```

**Step 1: Mount Google Drive**

Google Drive is mounted to access the dataset and save the trained model.

```python
from google.colab import drive

drive.mount('/content/drive')
```

**Step 2: Load the Dataset**

Load the Employee Salary Classification dataset from Google Drive.

```python
df = pd.read_excel(
    "/content/drive/MyDrive/Datasets/Salary_Logistic_Regression_Binary_Classification_Dataset.xlsx"
)

df
```

**Step 3: Handle Missing Values**

Check whether the dataset contains missing values.

If any missing values are found, replace them with the median value of the respective column.

```python
print("\nBefore Filling Missing Values")
print(df.isnull().sum())

df["Education_Level"] = df["Education_Level"].fillna(
    df["Education_Level"].median()
)

df["Age"] = df["Age"].fillna(
    df["Age"].median()
)

print("\nAfter Filling Missing Values")

print(df.isnull().sum())
```

**Step 4: Define Features and Target Variable**

Independent Variables (X):

- Experience_Years
- Education_Level
- Age

Dependent Variable (Y):

- High Salary

```python
X = df[
    [
        "Experience_Years",
        "Education_Level",
        "Age"
    ]
]

y = df["High_Salary"]

print(X.head())
print(y.head())
```

**Step 5: Split the Dataset**

Split the dataset into training and testing sets.

- 80% Training Data
- 20% Testing Data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

**Step 6: Scale the Features**

Scale the numerical features using **StandardScaler** so they have a similar scale.

StandardScaler applies the following formula to each feature:

\[
z=\frac{x-\mu}{\sigma}
\]

Where:

- **x** = Original value
- **μ** = Mean of the feature
- **σ** = Standard deviation of the feature
- **z** = Standardized value

Feature scaling helps Logistic Regression learn more effectively by preventing features with larger values from having a greater influence on the model.

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)

print(X_test)
```

**Step 7: Train the Logistic Regression Model**

Train the Logistic Regression model using the training dataset.

```python
model = LogisticRegression()

model.fit(X_train, y_train)

```

**Step 8: Make Predictions**

Predict the High Salary category using the testing dataset.

```python
y_pred = model.predict(X_test)

print(y_pred)
```

**Step 9: Calculate Accuracy**

Measure the percentage of correct predictions made by the model.

```python
accuracy = accuracy_score(
    y_test,
    y_pred
)

print("Accuracy:")

print(round(accuracy * 100,2),"%")
```

**Step 10: Display the Confusion Matrix**

A **Confusion Matrix** is a table used to evaluate the performance of a classification model by comparing the **actual values** with the **predicted values**.

It summarizes the model's predictions into four possible outcomes:

- **True Positive (TP):** The employee actually belongs to the **High Salary** category, and the model correctly predicts **High Salary**.
- **True Negative (TN):** The employee actually belongs to the **Normal Salary** category, and the model correctly predicts **Normal Salary**.
- **False Positive (FP):** The employee actually belongs to the **Normal Salary** category, but the model incorrectly predicts **High Salary**.
- **False Negative (FN):** The employee actually belongs to the **High Salary** category, but the model incorrectly predicts **Normal Salary**.

The confusion matrix helps identify where the model makes correct predictions and where it makes classification errors.

```python
cm = confusion_matrix(y_test,y_pred)

print(cm)
```

**Step 11: Display the Classification Report**

The classification report provides detailed performance metrics for the classification model.

It includes:

- **Precision:** When the model predicts **"High Salary,"** how often is it correct?
- **Recall:** Out of all actual **"High Salary"** employees, how many did the model correctly identify?
- **F1-Score:** A balanced measure that combines Precision and Recall into a single performance metric.
- **Support:** The number of actual observations belonging to each class in the testing dataset.

> **Note:** Accuracy is calculated separately in the previous step because it measures the model's overall correctness, whereas the classification report focuses on evaluating each class individually.

```python

print(classification_report(y_test, y_pred))

```

**Step 12: Predict the Salary Category for a New Employee**

Predict whether a new employee belongs to the High Salary category.

Employee Details:

- Experience = 10 Years
- Education Level = 18 Years
- Age = 35 Years

```python
new_employee = scaler.transform(
    [[10, 18, 35]]
)

prediction = model.predict(
    new_employee
)

print("Prediction:")

if prediction[0] == 1:
    print("High Salary")
else:
    print("Normal Salary")
```

The trained model was then used to classify employees into High Salary and Normal Salary categories. Finally, the model's performance was evaluated using accuracy, a confusion matrix, and a classification report, demonstrating the complete workflow for building and assessing a binary classification model.
