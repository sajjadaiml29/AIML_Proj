# K-Fold Cross Validation

## Objective

The objective of this notebook is to evaluate a Machine Learning model using K-Fold Cross Validation.

In this notebook, you will learn how to:

- Load the dataset
- Handle missing values
- Define features and target variable
- Create a Machine Learning pipeline
- Perform K-Fold Cross Validation
- Calculate the accuracy for each fold
- Calculate the average model accuracy

```python
# Import Required Libraries

import numpy as np
import pandas as pd

from sklearn.model_selection import (
    KFold,
    cross_val_score
)

from sklearn.preprocessing import StandardScaler

from sklearn.pipeline import Pipeline

from sklearn.linear_model import LogisticRegression
```

**Step 1: Mount Google Drive**

Google Drive is mounted to access the dataset and save the trained model.

```python
from google.colab import drive

drive.mount('/content/drive')
```

**Step 2: Load the Dataset**

Load the Employee Salary Classification dataset from Google Drive.

```python
df = pd.read_excel(
    "/content/drive/MyDrive/Datasets/Salary_Logistic_Regression_Multiclass_Classification_Dataset.xlsx"
)

df
```

## Step 3: Handle Missing Values

Check whether the dataset contains missing values.

If any missing values are found, replace them with the median value of the respective column.

```python
print("\nBefore Filling Missing Values")
print(df.isnull().sum())

df["Education_Level"] = df["Education_Level"].fillna(
    df["Education_Level"].median()
)

df["Age"] = df["Age"].fillna(
    df["Age"].median()
)

print("\nAfter Filling Missing Values")

print(df.isnull().sum())
```

## Step 4: Define Features and Target Variable

The independent variables are:

- Experience Years
- Education Level
- Age

The dependent variable is:

- Salary Category

Salary Categories:

- 0 = Low Salary
- 1 = Medium Salary
- 2 = High Salary
- 3 = Executive Salary

```python
X = df[
    [
        "Experience_Years",
        "Education_Level",
        "Age"
    ]
]

y = df["Salary_Category"]
```

## Step 5: Create a Machine Learning Pipeline

Create a pipeline that first scales the features using **StandardScaler** and then trains a **Logistic Regression** model.

Using a pipeline ensures that all preprocessing steps are applied correctly during each fold of Cross Validation.

```python
pipeline = Pipeline(
    [
        (
            "scaler",
            StandardScaler()
        ),
        (
            "model",
            LogisticRegression(
                max_iter=500
            )
        )
    ]
)
```

## Step 6: Create the K-Fold Cross Validator

Create a K-Fold Cross Validator with:

- 5 Folds
- Shuffled Data
- Random State = 42

Each fold will be used once as the testing dataset while the remaining folds are used for training.

```python
kf = KFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)
```

## Step 7: Perform K-Fold Cross Validation

Evaluate the Logistic Regression model using the K-Fold Cross Validator.

The accuracy score will be calculated for each fold.

```python
scores = cross_val_score(
    pipeline,
    X,
    y,
    cv=kf,
    scoring="accuracy"
)
```

## Step 8: Display the Accuracy for Each Fold

Display the accuracy obtained from each fold during K-Fold Cross Validation.

Each value represents the model's accuracy when one fold is used for testing and the remaining folds are used for training.

```python
print("Accuracy for Each Fold:")

print(scores)
```

## Step 9: Calculate the Average Accuracy

Calculate the average accuracy across all folds.

The average accuracy provides a more reliable estimate of the model's overall performance.

```python
average_accuracy = scores.mean()

print("Average Accuracy:")

print(
    round(average_accuracy * 100,2),
    "%"
)
```

## Step 10: Display the Accuracy for Each Fold and the Average Accuracy

Display the accuracy for each fold along with the average accuracy to evaluate the model's overall performance.

```python
print("Accuracy for Each Fold:")

for i, score in enumerate(scores, start=1):

    print(
        f"Fold {i}:",
        round(score * 100,2),
        "%"
    )

print()

print(
    "Average Accuracy:",
    round(scores.mean() * 100,2),
    "%"
)
```

## Step 11: Calculate the Standard Deviation of the Accuracy Scores

Calculate the standard deviation of the accuracy scores obtained from all folds.

A smaller standard deviation indicates that the model performs consistently across different folds.

```python
std_accuracy = scores.std()

print("Standard Deviation:")

print(
    round(std_accuracy * 100,2),
    "%"
)
```

## Step 12: Interpret the Results

Interpret the Cross Validation results.

- High average accuracy indicates good overall model performance.
- A low standard deviation indicates consistent performance across all folds.
- Large differences between fold accuracies may indicate that the model is sensitive to different subsets of the data.

```python
print("Model Evaluation Summary")

print("------------------------")

print(
    "Average Accuracy :",
    round(scores.mean() * 100,2),
    "%"
)

print(
    "Standard Deviation :",
    round(scores.std() * 100,2),
    "%"
)
```
